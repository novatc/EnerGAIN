```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('../data/clean/dataset_01102018_01012023.csv')

# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Set the 'date' column as the index of the DataFrame
df.set_index('date', inplace=True)

# Generate plots for each column
for column in df.columns:
    df[column].plot()
    plt.title(column)
    plt.show()

# Comparing multiple data series in one graph
# For example, let's say you want to compare 'Eingespeiste Energie kWh', 'prediction', 'prize', and 'consumption' in the same graph
df[['Eingespeiste Energie kWh', 'prediction', 'prize', 'consumption']].plot()
plt.title('Comparison graph')
plt.show()
```

```python
# Plotting the prediction and the actual values for the last 300 data points
df[['prediction', 'consumption']].tail(1000).plot()
```

```python
# multiple 'Eingespeiste Energie kWh' and 'prize' and safe it as a new column

df['potential'] = df['Eingespeiste Energie kWh'] * df['prize']/1000
df['potential'].plot()

```
